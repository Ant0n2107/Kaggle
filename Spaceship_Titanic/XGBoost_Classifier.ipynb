# Преобразование категориальных признаков в численные

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
train_2 = pd.read_csv('train.csv')
test_2 = pd.read_csv('test.csv')
```

```python
train_2['HomePlanet'] = train_2['HomePlanet'].fillna('Not stated')
train_2['CryoSleep'] = train_2['CryoSleep'].fillna(0)
train_2['Cabin'] = train_2['Cabin'].fillna('Not stated')
train_2['Destination'] = train_2['Destination'].fillna('Not stated')
train_2['VIP'] = train_2['VIP'].fillna(0)
train_2['Name'] = train_2['Name'].fillna('Not stated')
train_2['Age'] = train_2['Age'].fillna(0)
train_2['RoomService'] = train_2['RoomService'].fillna(0)
train_2['FoodCourt'] = train_2['FoodCourt'].fillna(0)
train_2['ShoppingMall'] = train_2['ShoppingMall'].fillna(0)
train_2['Spa'] = train_2['Spa'].fillna(0)
train_2['VRDeck'] = train_2['VRDeck'].fillna(0)
```

```python
test_2['HomePlanet'] = test_2['HomePlanet'].fillna('Not stated')
test_2['CryoSleep'] = test_2['CryoSleep'].fillna(0)
test_2['Cabin'] = test_2['Cabin'].fillna('Not stated')
test_2['Destination'] = test_2['Destination'].fillna('Not stated')
test_2['VIP'] = test_2['VIP'].fillna(0)
test_2['Name'] = test_2['Name'].fillna('Not stated')
test_2['Age'] = test_2['Age'].fillna(0)
test_2['RoomService'] = test_2['RoomService'].fillna(0)
test_2['FoodCourt'] = test_2['FoodCourt'].fillna(0)
test_2['ShoppingMall'] = test_2['ShoppingMall'].fillna(0)
test_2['Spa'] = test_2['Spa'].fillna(0)
test_2['VRDeck'] = test_2['VRDeck'].fillna(0)
```

```python
train_2['CryoSleep'] = train_2['CryoSleep'].astype(int)
train_2['VIP'] = train_2['VIP'].astype(int)
train_2['Age'] = train_2['Age'].astype(int)
train_2['RoomService'] = train_2['RoomService'].astype(int)
train_2['FoodCourt'] = train_2['FoodCourt'].astype(int)
train_2['ShoppingMall'] = train_2['ShoppingMall'].astype(int)
train_2['Spa'] = train_2['Spa'].astype(int)
train_2['VRDeck'] = train_2['VRDeck'].astype(int)
train_2['Transported'] = train_2['Transported'].astype(int)
```

```python
test_2['CryoSleep'] = test_2['CryoSleep'].astype(int)
test_2['VIP'] = test_2['VIP'].astype(int)
test_2['Age'] = test_2['Age'].astype(int)
test_2['RoomService'] = test_2['RoomService'].astype(int)
test_2['FoodCourt'] = test_2['FoodCourt'].astype(int)
test_2['ShoppingMall'] = test_2['ShoppingMall'].astype(int)
test_2['Spa'] = test_2['Spa'].astype(int)
test_2['VRDeck'] = test_2['VRDeck'].astype(int)
```
